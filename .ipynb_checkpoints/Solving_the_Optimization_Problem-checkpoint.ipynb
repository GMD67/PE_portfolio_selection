### Optimization Model

```python
# Current round stopping time tau 
# Round 1
Round = 1 + 1
```

```python
# Define yy
yy0 = np.ones((K,Round,I))
# Flatten
yy_flatten = np.reshape(yy0,-1)
# xixi same size with yy_flatten
xixi_flatten = np.reshape(xixi[:,:Round,:],-1)
```

```python
# Max return 
def objective(yy_flatten,xixi_flatten,R,K,I,Round,fund):
    beta = []
    for k in range(K-1):
        sum_return = np.sum(yy_flatten[k*Round*I:(k+1)*Round*I]*xixi_flatten[k*Round*I:(k+1)*Round*I]) - fund*Round
        if sum_return>= R:
            beta.append(1)
        else:
            beta.append(0)
    return -sum(beta)
```

```python
#test
objective(yy_flatten,xixi_flatten,R,K,I,Round,F)
```

```python
# Group of Constraints 
# start with rounds>=2
# def constraint1(yy_flatten,xixi_flatten,k,i,t):
#     return yy_flatten[(t-2)*k*i:(t-1)*k*i]*yy_flatten[(t-1)*k*i:t*k*i] - yy_flatten[(t-1)*k*i:t*k*i]
```

##### Simple Optimize

```python

# optimize

# Y range
b = (0.0,1.0)
bnds = ((b,)*K*Round*I)

# Optimization
solution = minimize(objective,yy_flatten,args=(xixi_flatten,R,K,I,Round,F),method='SLSQP',bounds=bnds)
y = solution.x

# show final objective
print('Final SSE Objective: ' + str(objective(y,xixi_flatten,R,K,I,Round,F)))
```
