```python
# Parameters generated by random guess

xixi = np.random.power(3.0,(K,I,T))-0.5
unicorn_index = np.random.randint(I,size = unicorn_size)
```

### Optimization Model

```python
# Define yy
yy = np.ones((K,I,T))
```

```python
#betabeta start from t = 1

for t in range(1,T):
    beta = []
    for k in range(K):
        for i in range(I):
            
            # Investment stopping conditions
            if np.sum(xixi[k,i,:t]) < -1 or yy[k,i,t-1] == 0:
                yy[k,i,t] = 0
                print('stop investing company',i,'\n')
02
    # Calculate beta
        result = np.sum(np.multiply(yy[k,:,t],xixi[k,:,t])) - F*t 
        if result >= R:
            beta.append(1)
        else:
            beta.append(0)
     
    # Check betabeta condition
    if sum(beta) >= p*K:
        print('optimal rounds =',t,'\n')
        yy_sum = np.sum(yy,axis=0)
        break
    else:
        print('rounds',t,'contiune \n')
            
```

```python
def cal_beta(yy,xixi,t):
    beta = []
    for k in range(K):
        result = np.sum(np.multiply(yy[k,:,t],xixi[k,:,t])) - F*t 
        if result >= R:
            beta.append(1)
        else:
            beta.append(0)
    return beta
```
